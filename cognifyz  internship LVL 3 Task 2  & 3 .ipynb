# Importing Required Libraries

```python
import numpy as np
import pandas as pd 
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Loading Dataset

```python
df=pd.read_csv('cognifyz Dataset .csv')
df
```

# Task 2 : Customer Preference Analysis

1. Analyze the relationship between the type of cuisine and the restaurant's rating.


2. Identify the most popular cuisines among customers based on the number of votes.


3. Determine if there are any specific cuisines that tend to receive higher ratings.

##### 1. Analyze the relationship between the type of cuisine and the restaurant's rating.¶

```python
cuisine_rating = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
cuisine_rating.head(30)
```

```python
cuisine_rating = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
```

```python
# Extract the top 10 cuisines from the cuisine_rating DataFrame
best_cuisines = cuisine_rating.head(20)

# Set up the figure and axes with a defined size
plt.figure(figsize=(15, 8))

# Visualize the top cuisines using a bar plot
ax = sns.barplot(x=best_cuisines.index, y=best_cuisines.values, palette='Set2')

# Display labels with values on top of the bars
for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}', (bar.get_x() + bar.get_width() / 2., height), ha='center', va='bottom', fontsize=10)

# Customize the plot appearance
plt.title('Top 20 Cuisines by Average Rating', fontsize=18)
plt.xlabel('Cuisine', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.xticks(rotation=90, fontsize=11)

# Display the plot
plt.tight_layout()
plt.show()
```

##### 2. Identify the most popular cuisines among customers based on the number of votes.

```python
cuisine_votes = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False)
cuisine_votes.head(10)
```

```python
# Sort the cuisine_votes DataFrame and select the top 10 cuisines
top_cuisines = cuisine_votes.head(10)

# Create a figure and axes with specified size
plt.figure(figsize=(12, 6))

# Create a bar plot
ax = sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette='viridis')

# Add labels with values on top of the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height):,}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', fontsize=10)

# Customize the plot
plt.title('Top 10 Cuisines by Number of Votes', fontsize=18)
plt.xlabel('Cuisine', fontsize=12)
plt.ylabel('Total Votes', fontsize=12)
plt.xticks(rotation=35, fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

##### 3. Determine if there are any specific cuisines that tend to receive higher ratings.

```python
# Group by 'Cuisines' and calculate the correlation between 'Rating text' and 'Votes'
correlation_cuisine_rating_votes = df.groupby('Cuisines')[['Rating text', 'Votes']].corr().iloc[0::2, -1]

# Filter cuisines with a positive correlation
higher_rated_cuisines = correlation_cuisine_rating_votes[correlation_cuisine_rating_votes > 0].index.get_level_values(0)

# Print the result
print(f"Cuisines that tend to receive higher ratings: {', '.join(higher_rated_cuisines)}")
```

##### 1. Create visualizations to represent the distribution of ratings using different charts (histogram, barplot, etc.)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=30, kde=True)
plt.title('Distribution of Ratings (Histogram)')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
rating_counts = df['Aggregate rating'].value_counts().sort_index()
sns.barplot(x=rating_counts.index, y=rating_counts.values, color='lightgreen')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings (Bar Plot)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['Aggregate rating'], color= '#e74c3c')
plt.title('Distribution of Ratings (Violin Plot)')
plt.xlabel('Aggregate Rating')
plt.show()
```

##### 2. Compare the average ratings of different cuisines or cities using appropriate visualizations

```python
def plot_average_ratings(data, x_label, y_label, title, color):
    plt.figure(figsize=(12, 6))
    sns.barplot(x=data.index, y=data.values, color=color)
    plt.xticks(rotation=90)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()

# Average Ratings by Cuisine
avg_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(30)
plot_average_ratings(avg_rating_by_cuisine, 'Cuisines', 'Average Rating', 'Average Ratings by Cuisine', '#8e44ad')

# Average Ratings by City
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(30)
plot_average_ratings(avg_rating_by_city, 'City', 'Average Rating', 'Average Ratings by City', '#3498db')
```

##### 3. Visualize the relationship between various features and the target variable to gain insights.

```python
avg_rating_by_price = df.groupby('Price range')['Aggregate rating'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Price range', y='Aggregate rating', data=avg_rating_by_price,color='purple')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Create a scatterplot with enhanced aesthetics
ax = sns.scatterplot(x='Votes', y='Aggregate rating', data=df, color='skyblue')

# Customize the plot labels
plt.title('Votes vs. Aggregate Rating', fontsize=16)
plt.xlabel('Votes', fontsize=12)
plt.ylabel('Aggregate Rating', fontsize=12)

# Add gridlines for reference
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Count the number of restaurants with and without online delivery
online_delivery_count = df['Has Online delivery'].value_counts()

# Define data for the pie chart
labels = ['No Online Delivery', 'Online Delivery']
colors = ['#ff9999', '#66cc99']  # Updated colors
explode = (0.1, 0)  # Explode the first slice (No Online Delivery)

# Create a figure with a specified size
plt.figure(figsize=(10, 10))

# Create a pie chart with enhanced aesthetics
plt.pie(online_delivery_count, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)

# Customize the plot title
plt.title('Restaurant Online Delivery Availability', fontsize=16)

# Ensure an equal aspect ratio for a circular pie chart
plt.axis('equal')

# Show the plot
plt.show()
```

```python
# Count the number of restaurants with and without table booking
table_booking_count = df['Has Table booking'].value_counts()

# Define data for the pie chart
labels = ['No Table Booking', 'Table Booking']
colors = ['#ff6666', '#99ccff']  # Updated colors
explode = (0.1, 0)  # Explode the first slice (No Table Booking)

# Create a figure with a specified size
plt.figure(figsize=(8, 8))

# Create a pie chart with enhanced aesthetics and custom colors
plt.pie(table_booking_count, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)

# Customize the plot title
plt.title('Restaurant Table Booking Availability', fontsize=16)

# Ensure an equal aspect ratio for a circular pie chart
plt.axis('equal')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Create a scatterplot with enhanced aesthetics and symlog scale on the x-axis
ax = sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=df, color='green')

# Customize the plot labels
plt.title('Average Cost for Two vs. Aggregate Rating', fontsize=16)
plt.xlabel('Average Cost for Two (Symlog Scale)', fontsize=12)
plt.ylabel('Aggregate Rating', fontsize=11)

# Add gridlines for reference
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Top 10 city by restaurant count 

city_restaurant_count = df['City'].value_counts()[:10]

# Create a figure and axes with specified size
plt.figure(figsize=(12, 6))

# Sort the city_restaurant_count Series in descending order
sorted_city_restaurant_count = city_restaurant_count.sort_values(ascending=False).head(10)

# Create a bar plot with enhanced aesthetics
ax = sns.barplot(x=sorted_city_restaurant_count.index, y=sorted_city_restaurant_count.values, palette='viridis')

# Set log scale on the y-axis
ax.set_yscale('log')

# Customize the plot labels and rotation of x-axis labels
plt.title('Top 10 Cities by Restaurant Count', fontsize=16)
plt.xlabel('City', fontsize=11)
plt.ylabel('Number of Restaurants', fontsize=11)
plt.xticks(rotation=45, ha='right')

# Add labels with values on top of the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
